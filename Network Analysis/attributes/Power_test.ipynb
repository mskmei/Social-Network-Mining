```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
G = nx.read_edgelist('data/facebook_combined.txt', delimiter=' ', nodetype=int)
```

```python
degree = []
counts = list(nx.degree_histogram(G))
for i in range(len(counts)):
    degree.append(i)
log_degree = []
log_counts = []
degree_nonzero = []
counts_nonzero = []
for i in range(len(degree)):
    if (counts[i]):
        degree_nonzero.append(degree[i])
        counts_nonzero.append(counts[i])
        log_degree.append(np.log(degree[i]))
        log_counts.append(np.log(counts[i]))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_title("degree-counts based on ego-Facebook")
plt.xlabel("degree")
plt.ylabel("counts")
plt.scatter(degree_nonzero, counts_nonzero, c=counts_nonzero)
plt.savefig("degree-counts.png")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_title("log(degree)-log(counts) based on ego-Facebook")
plt.xlabel("log(degree)")
plt.ylabel("log(counts)")
plt.scatter(log_degree, log_counts, c=log_counts)
plt.savefig("log-degree-counts.png")
plt.show()
```
